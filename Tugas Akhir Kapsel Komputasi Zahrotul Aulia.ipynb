```python
# Importing Modules untuk Notebook ini
import warnings; warnings.simplefilter('ignore')
import pandas as pd, numpy as np, matplotlib.pyplot as plt, seaborn as sns
from scipy.stats.stats import pearsonr
```

```python
d = pd.read_csv('ramen-ratings.csv')
B, K = d.shape
print('baris = ', B, ', Kolom = ', K)
d.head(10)
```

### Data Wrangling & Preprocessing

```python
# Cek type data
d.dtypes
```

```python
# Mengubah type data Stars dari object menjadi numerik(float64)
d['Stars'] = pd.to_numeric(d['Stars'], errors='coerce')
```

```python
# Cek type data setelah diubah
d.dtypes
```

```python
# Cek apakah terdapat missing value
d.describe(include='all')
```

```python
# Mengganti nilai \n pada variabel diatas dengan nilai NaN
d[['Top Ten']] = d[['Top Ten']].replace('\n',np.nan)
```

```python
# Menghitung jumlah data missing
d.isnull().sum()
```

Jika dilihat variabel Style terdapat missing value sebanyak 2 baris, variabel Stars sebanyak 3 baris, dan variabel Top Ten sebanyak 2543 baris. Sehingga perlu dihapus variabel Top Ten karena terjadi missing value yang lebih dari 50% dataset yang dimiliki, dan perlu dihapus baris yang terdapat missing value pada variabel Style dan Stars.

```python
# Menghapus data missing variabel Top Ten
d.drop("Top Ten", axis=1, inplace=True)
```

```python
# Menghapus baris yang terdapat data missing
dc = d.dropna()
```

```python
print(dc.shape)
dc
```

```python
# Cek apakah terdapat outlier
# Histogram dan Scatter Plot
sns.pairplot(dc)
```

```python
# Trimming : IQR
Q1 = dc.quantile(0.25)
Q3 = dc.quantile(0.75)
IQR = Q3 - Q1
print(IQR) # Menghitung jumlah data setelah di trim dengan IQR
```

```python
dataramen = dc[~((dc<(Q1-1.5*IQR))|(dc>(Q3+1.5*IQR))).any(axis=1)]
dataramen.shape # data setelah di cek outlier
```

```python
dc.shape # data sebelum cek outlier
```

Jadi, data yang merupakan outlier dengan kata lain jauh dari data-data yang lain telah dibuang sejumlah 129 dengan IQR dari dataset

#### Apakah jumlah Review berpengaruh signifikan terhadap Ratings/Stars ?

```python
# Koefisien Korelasi
x = dataramen['Review #']
y = dataramen['Stars']
pearsonr(x,y)
```

Dilihat dari hasil koefisien korelasi dengan nilai 0.31 dapat dikatakan nilainya rendah, artinya rating atau variabel stars tidak dipengaruhi oleh jumlah review atau variabel review.
